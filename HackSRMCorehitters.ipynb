```python
# read the data into a Pandas DataFrame
import pandas as pd

url = 'datasheets.csv'

col_names = ['Time','heart Rate','label']

pima = pd.read_csv(url, header=None, names=col_names)
```

```python
# print the first 5 rows of data from the dataframe
i=pima[98:]
i.head()
```

```python
# define X and y
feature_cols = ['heart Rate']

# X is a matrix, hence we use [] to access the features we want in feature_cols
X = i[feature_cols]

# y is a vector, hence we use dot to access 'label'
z = i.label
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, z_train, z_test = train_test_split(X, z, random_state=0)
```

```python
# train a logistic regression model on the training set
from sklearn.linear_model import LogisticRegression
# instantiate model
logreg = LogisticRegression()

# fit model
logreg.fit(X_train,z_train)
```

```python
# make class predictions for the testing set
z_pred_class = logreg.predict(X_test)
#LogisticRegression(solver='lbfgs')
```

```python
# calculate accuracy
from sklearn import metrics
print(metrics.accuracy_score(z_test, z_pred_class)*100,"%")
```
